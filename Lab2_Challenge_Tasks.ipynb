```python
target_name  = "MABALOT"
SEED_NUM = "5"
redacted_name = ""

for x in target_name:
    if x in ["A", "E", "I", "O", "U"]:
        redacted_name += SEED_NUM
    else:
        redacted_name += x

print("Redacted name:", redacted_name)
```

```python
ENGINEER = "Mabalot"
STUDENT_ID = "TUPM-25-0705"

SEED_NUM = int(STUDENT_ID[-1])

first_letter = ENGINEER[0]
ASCII = ord(first_letter)

loops = SEED_NUM + 3

A = ASCII
B = SEED_NUM
C = loops

verification_hash = (A * B) - C

print("First Letter ASCII (A):", A)
print("SEED_NUM (B):", B)
print("Total Loops (C):", C)
print(f"Forensic Signature: {verification_hash}")
```
